```python
dir1 = '/content/drive/MyDrive/Crop___DIsease/Corn___Common_Rust'
dir2 = '/content/drive/MyDrive/Crop___DIsease/Corn___Healthy'
dir3 = '/content/drive/MyDrive/Crop___DIsease/Corn___Leaf_Blight'
dir4 = '/content/drive/MyDrive/Crop___DIsease/Corn___Gray_Leaf_Spot'
```

```python
for i in os.listdir(dir1)[:500]:
    img = plt.imread(dir1+"/"+i)
    img = cv2.resize(img,(200,200))
    sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
    newarr = img.reshape(img.shape[0], (img.shape[1]*img.shape[2]))
    low_variance_thrusholing = sel.fit_transform(newarr)
    pil_image  = Image.fromarray(low_variance_thrusholing).convert('RGB')
    open_cv_image = np.array(pil_image)
    open_cv_image = cv2.resize(open_cv_image,(200,200))
    edges = cv2.Canny(open_cv_image, t_lower, t_upper)
    images.append(edges.flatten())
    class_.append(1)
```

```python
for i in os.listdir(dir2)[:500]:
    img = plt.imread(dir2+"/"+i)
    img = cv2.resize(img,(200,200))

    sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
    newarr = img.reshape(img.shape[0], (img.shape[1]*img.shape[2]))
    low_variance_thrusholing = sel.fit_transform(newarr)
    pil_image  = Image.fromarray(low_variance_thrusholing).convert('RGB')
    open_cv_image = np.array(pil_image)
    open_cv_image = cv2.resize(open_cv_image,(200,200))
    edges = cv2.Canny(open_cv_image, t_lower, t_upper)
    images.append(edges.flatten())
    class_.append(2)
```

```python
for i in os.listdir(dir3)[:500]:
    img = plt.imread(dir3+"/"+i)
    img = cv2.resize(img,(200,200))

    sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
    newarr = img.reshape(img.shape[0], (img.shape[1]*img.shape[2]))
    low_variance_thrusholing = sel.fit_transform(newarr)
    pil_image  = Image.fromarray(low_variance_thrusholing).convert('RGB')
    open_cv_image = np.array(pil_image)
    open_cv_image = cv2.resize(open_cv_image,(200,200))
    edges = cv2.Canny(open_cv_image, t_lower, t_upper)
    images.append(edges.flatten())
    class_.append(3)
```

```python
for i in os.listdir(dir4)[:500]:
    img = plt.imread(dir4+"/"+i)
    img = cv2.resize(img,(200,200))

    sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
    newarr = img.reshape(img.shape[0], (img.shape[1]*img.shape[2]))
    low_variance_thrusholing = sel.fit_transform(newarr)
    pil_image  = Image.fromarray(low_variance_thrusholing).convert('RGB')
    open_cv_image = np.array(pil_image)
    open_cv_image = cv2.resize(open_cv_image,(200,200))
    edges = cv2.Canny(open_cv_image, t_lower, t_upper)
    images.append(edges.flatten())
    class_.append(4)
```

```python
images = np.asarray(images)
class_ = np.asarray(class_)
print(images.shape)
print(class_.shape)
from sklearn.model_selection import train_test_split
X_train, X_test ,Y_train, Y_test = train_test_split(images,class_,test_size= 0.2, shuffle=True)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
kernel = 1.0 * RBF(1.0)

gpc = GaussianProcessClassifier(kernel=kernel,random_state=0).fit(X_train, Y_train)
Gaussianpred = gpc.predict(X_test)

acc6 = metrics.accuracy_score(Y_test,Gaussianpred)
```
